```python
PATH_TO_AS7263_DATA = '../HarvardMuseumData/IlluminationSources/'
SPECTRASUITE_DATA_FILE = '../SpectraSuiteData/HalogenReferenceSpectrum.txt'
BAND = np.array([610,680,730,760,810,860])
```

```python
import csv
import numpy as np
import pandas as pd
import os 
import sklearn.preprocessing as prep

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def CSVtoNumpy(csvFile):
    df=pd.read_csv(csvFile, sep=',',header=None)
    return df.values
```

```python
colors = ['red','blue','green','orange','lime','fuchsia','black','yellow','turquoise']

def PlotIlluminationSources(path): 
    for i,csvFile in enumerate(os.listdir(path)): 
        #skip the DS_Store directory (auto-generated by Dropbox(?))
        if(csvFile == '.DS_Store'): 
            continue 
        else:
            key = csvFile.split('.')[0]
            if key == 'MITPointBulbatSensorClose':
                continue
            elif key == 'MITPointBulbatSensorFar2':
                raw_data = CSVtoNumpy(path + csvFile)
                raw_data = np.delete(raw_data, -1, 1)
                print 'key: ' + str(key) + ' color:' + colors[i]
                plt.plot(BAND,np.transpose(raw_data),color=colors[i],linewidth='3')
```

```python
PlotIlluminationSources(PATH_TO_AS7263_DATA)
```

```python
def plotSpectraSuiteData():
    spectrasuite = open(SPECTRASUITE_DATA_FILE,'r')
    bands = []
    values = []
    vals = []
    data = spectrasuite.read().split('\n')
    for row in data: 
        vals.append([float(a) for a in row.split('\t')])
    #print vals
    for val in vals: 
        bands.append(val[0])
        values.append(val[1])       

    maxval = np.max(values)
    normalized_values = [a / maxval for a in values]
    print bands[1910]
    print normalized_values[1910]
    #added_bands = [a+100 for a in bands]
    plt.plot(bands,np.transpose(normalized_values))
    
```
